```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from imblearn.combine import SMOTEENN
import pickle
```

```python
#reading the csv file

df= pd.read_csv('./processed_data.csv')
df.head()
```

```python
df =df.drop(columns= ['Unnamed: 0'])
df
```

As we discussed in EDA that the dataset is imbalanced so we need to make the dataset balanced before proceeding forward

```python
#Now we have properly processed numerical data present with us
x= df.drop(columns= ['Churn']) #input
y= df['Churn'] #output
x
```

```python

#BALANCING THE DATASET- unbalanced dataset will give low accuracy scores
sm= SMOTEENN()
x_resampled, y_resampled= sm.fit_resample(x, y)
x_resampled
```

```python
#split into training and testing data 

xr_train,xr_test,yr_train,yr_test = train_test_split(x_resampled,y_resampled, test_size=0.2, random_state=5)
```

### DECISION TREE CLASSIFIER

```python
model_dt= DecisionTreeClassifier(criterion='gini', random_state=100, max_depth=6 , min_samples_leaf=8)
```

```python
model_dt.fit(xr_train, yr_train)
```

```python
y_pred= model_dt.predict(xr_test)
y_pred
```

```python
print(confusion_matrix(yr_test, y_pred))
```

```python
print(classification_report(yr_test, y_pred, labels=[0,1]))
```

ACCURACY for decision tree classifier = (489+ 606)  / (489+ 606+ 37 + 33) = 1095 / 1165 = 0.9399 -> 94% (approx.)

### RANDOM FOREST CLASSIFIER

```python
model_rf= RandomForestClassifier(n_estimators=100, criterion='gini', random_state=100, max_depth=6 , min_samples_leaf=8)
```

```python
model_rf.fit(xr_train, yr_train)
```

```python
y_pred= model_rf.predict(xr_test)
```

```python
print(confusion_matrix(yr_test, y_pred))
```

```python
print(classification_report(yr_test, y_pred, labels=[0,1]))
```

ACCURACY for Random Forest Classifier = (459 + 636) / (459 + 636 + 58 + 2) = 1095 / 1155 = 0.948 -> 94.8% 

```python
filename= 'rf_model.sav'
pickle.dump(model_rf , open(filename, 'wb'))
```

```python
load_model= pickle.load(open(filename, 'rb'))
```

```python
load_model.score(xr_test, yr_test)
```
